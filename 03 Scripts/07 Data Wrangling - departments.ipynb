# 03. Transposing Data

```python
# 1. Transpose the 'dep' dataframe.
dep_t = dep.T
```

```python
dep_t
```

```python
# The column title above is not right. 
# 2. Create a new header for the new transposed data later.
new_header = dep_t.iloc[0]
```

```python
new_header
```

```python
# 3. Save a new version of the transposed data which include only the actual content (start from row 1, don't include row 0).
dep_t_new = dep_t.iloc[1:]
```

```python
dep_t_new
```

```python
# The new version of transposed data doesn't have a column header. Let's add them. 
# 4. Assign the new header we created in step 2 to this new version of transposed data. 
dep_t_new.columns = new_header
```

```python
dep_t_new
```

```python
# 5. Reset the index of the new transposed data. This will create a new column called 'index'.
dep_t_new = dep_t_new.reset_index()
```

```python
dep_t_new
```

```python
# 6. Change the 'index' column name to 'department_id'. 
# Don't worry if you see two identical column names here. It will disappear when we import this data back again. 
dep_t_new.rename(columns = {'index' : 'department_id'}, inplace = True)
```

```python
dep_t_new
```

# 04. Export Data

```python
# Export 'dep_t_new' to 'Prepared Data' folder in csv format
dep_t_new.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'departments_wrangled.csv'))
```
